```python
from google.colab import drive
drive.mount('/content/gdrive/')
import numpy as np      #행렬 데이터
import matplotlib.pyplot as plt
import scipy.io.wavfile
#import sounddevice as sd

v_samplerate, v_data = scipy.io.wavfile.read('/content/gdrive/MyDrive/Colab Notebooks/res/thank_you.wav')
times = np.arange(len(v_data)) / float(v_samplerate)

print('smpling rate', v_samplerate)
print('time : ', times[-1])
print('vData',v_data[5000:5100])

plt.plot(times, v_data)
plt.xlim(times[0], times[-1])
plt.xlabel('time (s)')
plt.ylabel('amplitude')
plt.show()
```
