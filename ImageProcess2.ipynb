```python
import numpy as np      #행렬 데이터
import matplotlib.pyplot as plt
import PIL.Image as pilimg

im = pilimg.open('/content/gdrive/MyDrive/Colab Notebooks/res/rgb_circle.bmp')

pix = np.array(im)
pixSize = np.array(pix.shape)
print(pixSize)

#원본 이미지에서 RGB에 해당하는 배열 만들기
pix_R = pix.copy()
pix_R[:,:,(1,2)] = 0
pix_G = pix.copy()
pix_G[:,:,(0,2)] = 0
pix_B = pix.copy()
pix_B[:,:,(0,1)] = 0

#원본 이미지인 pix 행렬을 이미지 데이터로 출력
plt.subplot(141)
plt.imshow(pix)
plt.axis('off')
plt.title('RGB')

#pix 행렬에서 이미지 데이터의 R채널 출력
plt.subplot(142)
plt.imshow(pix_R)
plt.axis('off')
plt.title('R')

#pix 행렬에서 이미지 데이터의 G채널 출력
plt.subplot(143)
plt.imshow(pix_G)
plt.axis('off')
plt.title('G')

#pix 행렬에서 이미지 데이터의 G채널 출력
plt.subplot(144)
plt.imshow(pix_B)
plt.axis('off')
plt.title('B')
plt.show()
```
